### Logistic regression

```python
def sigmoid(x : float):
    return 1 / (1 + np.exp(-x))

sigmoid(3)
```

```python
def binary_crossentropy(theta, X, y):
    h = sigmoid(X.dot(theta))
    m = X.shape[0]
    
    sum = 0
    for i in range(m):
        sum += y[i]*np.log(h[i]) + (1 - y[i])*np.log(1 - h[i])
        
    return sum / (-m)
    
```

```python
theta = (1, 1)
X = df.drop('admission', axis=1)
y = df['admission']
binary_crossentropy(theta, X, y)
```
